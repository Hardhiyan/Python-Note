# Basic Data Types


# Integers

Integer atau singkatnya "ints", adalah nilai numerik bilangan bulat. Setiap angka positif atau negatif (atau 0) tanpa desimal adalah bilangan bulat dalam Python. Nilai bilangan bulat memiliki presisi yang tidak terbatas, artinya bilangan bulat adalah eksak. Anda dapat memeriksa jenis objek Python dengan fungsi type(). Mari kita jalankan type() pada bilangan bulat:

```python
type(12)
```

Di atas kita melihat bahwa tipe 12 adalah tipe "int". Anda juga dapat menggunakan fungsi isinstance() untuk memeriksa apakah suatu objek adalah turunan dari tipe tertentu:

```python
# Check if 12 is an instance of type "int"

isinstance(12, int)
```

Output kode True menegaskan bahwa 12 adalah sebuah int.

Bilangan bulat mendukung semua operasi matematika dasar yang kita bahas terakhir kali. Jika operasi matematika yang melibatkan bilangan bulat akan menghasilkan nilai non-bilangan bulat (desimal), hasilnya menjadi float:

```python
1/3  # A third is not a whole number*
```

```python
type(1/3)  # So the type of the result is not an int
```

```python
type(4/2)
```

# Floats

Angka floating point atau "mengambang" adalah angka dengan nilai desimal. Tidak seperti bilangan bulat, bilangan floating point tidak memiliki presisi tak terbatas karena bilangan desimal irasional sangat panjang dan karenanya tidak dapat disimpan dalam memori. Sebagai gantinya, komputer memperkirakan nilai desimal panjang, jadi mungkin ada kesalahan pembulatan kecil dalam float panjang. Kesalahan ini sangat kecil sehingga biasanya tidak menjadi perhatian kami, tetapi dapat bertambah dalam kasus-kasus tertentu ketika membuat banyak perhitungan berulang.

Setiap angka dalam Python dengan titik desimal adalah float, bahkan jika tidak ada angka bukan nol setelah desimal:

```python
type(1.0)
```

```python
isinstance(0.33333, float)
```

Operasi aritmatika yang kita pelajari terakhir kali bekerja pada float dan juga int. Jika Anda menggunakan float dan int dalam ekspresi matematika yang sama, hasilnya adalah float:

```python
5 + 1.0
```

You can convert a float to an integer using the int() function:

```python
int(6.0)
```

You can convert an integer to a float with the float() function:

```python
float(6)
```

Floats can also take on a few special values: Inf, -Inf and NaN. Inf and -Inf stand for infinity and negative infinity respectively and NaN stands for "not a number", which is sometimes used as a placeholder for missing or erroneous numerical values.

```python
type ( float ("Inf") )
```

```python
type ( float ("NaN") )
```

*Note: Python contains a third, uncommon numeric data type "complex" which is used to store complex numbers.*

# Booleans

Boolean atau "bool" adalah nilai benar/salah yang dihasilkan dari pernyataan logis. Dalam Python, boolean dimulai dengan huruf pertama yang dikapitalisasi sehingga True dan False dikenali sebagai bool tetapi benar dan salah tidak. Kita telah melihat contoh boolean ketika kita menggunakan fungsi isinstance() di atas.

```python
type(True)
```

```python
isinstance(False, bool)  # Check if False is of type bool
```

You can create boolean values with logical expressions. Python supports all of the standard logic operators you'd expect:

```python
# Use >  and  < for greater than and less than:
    
20 > 10 
```

```python
20 < 5
```

```python
# Use >= and  <= for greater than or equal and less than or equal:

20 >= 20
```

```python
# Use == (two equal signs in a row) to check equality:

10 == 10
```

```python
40 == 40.0  # Equivalent ints and floats are considered equal
```

```python
# Use != to check inequality. (think of != as "not equal to")

1 != 2
```

```python
# Use the keyword "not" for negation:

not False
```

```python
# Use the keyword "and" for logical and:

(2 > 1) and (10 > 11)
```

```python
# Use the keyword "or" for logical or:

(2 > 1) or (10 > 11)
```

Mirip dengan ekspresi matematika, ekspresi logika memiliki urutan operasi yang tetap. Dalam pernyataan logis, perbandingan seperti >, < dan == dieksekusi terlebih dahulu, diikuti oleh "not", lalu "and" dan terakhir "or".

```python
2 > 1 or 10 < 8 and not True
```

```python
((2 > 1) or (10 < 8)) and (not True)
```

You can convert numbers into boolean values using the bool() function. All numbers other than 0 convert to True:

```python
bool(1)
```

```python
bool(0)
```

# Strings

Data teks dalam Python dikenal sebagai string atau "str". Mengelilingi teks dengan tanda kutip tunggal atau ganda untuk membuat string:

```python
type("cat")
```

```python
type('1')
```


Dua tanda kutip tepat di samping satu sama lain (seperti '' atau "") tanpa apa pun di antara mereka dikenal sebagai string kosong. String kosong sering kali mewakili nilai teks yang hilang.

# None

```python
type(None)  
```
