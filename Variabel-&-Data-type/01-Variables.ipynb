# Variables

Dalam pelajaran singkat ini, kita akan belajar bagaimana mendefinisikan variabel dengan Python. Variabel adalah nama yang Anda tetapkan nilai atau objek. Setelah menetapkan variabel, Anda dapat mengakses nilai atau objek terkaitnya menggunakan nama variabel. Variabel adalah cara yang nyaman untuk menyimpan nilai dengan nama yang bermakna.

Dalam Python, tetapkan variabel menggunakan `"="` :

```python
x = 10            
y = "Python is fun"         
z = 144**0.5 == 12

print(x)
print(y)
print(z)
```

*Catatan: menetapkan variabel tidak menghasilkan output apa pun.*

Saat menetapkan variabel, praktik yang baik adalah memberi spasi di antara nama variabel, operator penugasan, dan nilai variabel untuk kejelasan:

```python
p=8          # This works, but it looks messy.
print(p)   

p = 10       # Use spaces instead
print(p)
```

Setelah menetapkan variabel, Anda dapat melakukan operasi pada objek yang ditugaskan kepadanya menggunakan nama mereka:

```python
x + z + p
```

Anda dapat menetapkan objek yang sama ke beberapa variabel dengan pernyataan penugasan ganda

```python
n = m = 4
print(n)
print(m)
```

Anda juga dapat menetapkan beberapa variabel berbeda secara bersamaan menggunakan urutan nama variabel yang dipisahkan koma diikuti oleh operator penugasan dan urutan nilai yang dipisahkan koma di dalam tanda kurung:

```python
# Assign 3 variables at the same time:

x, y, z = (10, 20 ,30)

print(x)
print(y)
print(z)
```

Metode mengekstraksi variabel dari urutan yang dipisahkan koma ini dikenal sebagai `"membongkar tuple."`

Anda dapat menukar nilai dua variabel menggunakan sintaks yang serupa:

```python
(x, y) = (y, x)

print(x)
print(y)
```
